## Making the Recommender System

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()

def stem(text):
  y= []
  for i in text.split():
    y.append(ps.stem(i))
  return (" ".join(y))
```

```python
df1['tags']= df1['tags'].apply(stem)
df1['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features= 1000, stop_words= 'english')
```

```python
vectors= cv.fit_transform(df1['tags']).toarray()
vectors
```

```python
for i,j in zip(cv.get_feature_names_out(), range(1, 500)):
  print(i)
  if j == 500:
    break
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity= cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[17998])), reverse= True, key= lambda x: x[1])[:10]
```

```python
def recommend(book):
  ind= df1[df1['title'] == book].index
  sorted_sim= sorted(list(enumerate(similarity[ind.item()])), reverse= True, key= lambda x: x[1])[:10]
  for i, val in sorted_sim:
    print(df1.iloc[i]['title'])
recommend('Brothel: Mustang Ranch and Its Women')
```
